```python
import pandas as pd
file_path="D://python directory//data//input//"
file_name="GoogleAds_sales_data.csv"
```

```python
df_1=pd.read_csv(file_path+file_name)
```

```python
df_1["Cost"]=df_1["Cost"].str.replace("$","",regex=False).astype(float)
```

```python
df_1["Cost"]
```

```python
df_1["Sale_Amount"]=df_1["Sale_Amount"].str.replace("$","",regex=False).astype(float)
```

```python
df_1["Sale_Amount"]
```

```python
df_1["Ad_Date"] = pd.to_datetime(df_1["Ad_Date"], errors="coerce")
```

```python
df_1["Ad_Date"]
```

```python
df_1
```

```python
df_1.duplicated().sum()
```

```python
df_1.isnull().sum()
```

```python
df_1["Campaign_Name"]=df_1["Campaign_Name"].str.replace("Corse","course")
```

```python
df_1["Campaign_Name"]
```

```python
for i in df_1.select_dtypes(include="object"):
    df_1[i]=df_1[i].str.strip().str.lower()
```

```python
df_1["Ad_ID"]=df_1["Ad_ID"].str.upper()
```

```python
df_1
```

```python
df_1["Campaign_Name"]=df_1["Campaign_Name"].str.replace(" ","")
```

```python
df_1["Campaign_Name"]
```

```python
df_1
```

```python
df_1=df_1.select_dtypes(include=["int","float","object"]).fillna(0)
```

```python
df_1
```

```python
df_1.isnull().sum()
```
